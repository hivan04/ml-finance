An important caveat: dividends are no longer the whole payout story. Since the early
2000s, share repurchases (buybacks) have increasingly replaced dividends as the
dominant channel for returning cash to shareholders (Grullon & Michaely 2002b, Brav et
al. 2005). A company that holds its dividend flat while substantially increasing its buyback
programme is not necessarily signalling stagnation — it may simply prefer the flexibility
and tax efficiency of repurchases. This means that a model focused exclusively on
dividend changes will miss an important dimension of payout policy. You should be aware
of this limitation throughout the project. In the feature engineering phase (Task 1.3), you
will construct buyback-related features to help your model distinguish between genuine
dividend stagnation and substitution. In the discussion section of your report, you should
assess how much of your signal, if any, is contaminated by this substitution effect.

## Objective
Pipeline that:
(i) establishes the empirical existence of post-dividend-change return drift <br>
(ii) identifies the conditions under which this drift is strongest<br>
(iii) constructs a machine learning model to predict dividend changes using earnings-related features<br> 
(iv) translates these predictions into a dollar-neutral long-short equity strategy with realistic performance evaluation
